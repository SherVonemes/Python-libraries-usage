Берем тупо из дз:

```python
merged_data = pd.merge(orders, items, on="order_id")
merged_data = pd.merge(merged_data, sellers, on="seller_id")
merged_data = pd.merge(merged_data, customers, on="customer_id")

merged_data = merged_data.drop_duplicates(subset=['order_id', "seller_id"]) #вот это был мем, но логично, что если продавец и заказ повторяются, то нужно удалить

merged_data_full=merged_data.copy()

merged_data["in_other_state"] = (merged_data["seller_state"] != merged_data["customer_state"]) # сначала я считал наоборот, то есть доставку в свой штат,
#потом вычитал пропорпию из едницы, но у меня не проходили ассерты, пришлось поменять, хотя по памяти так было бы чуть быстрее; значения тру / фолс, но это и так понятно - переведем в инты
merged_data['in_other_state'] = merged_data['in_other_state'].astype(int)

counts = merged_data.groupby("seller_id")["order_id"].count().reset_index() #общая


other_state_counts = merged_data.groupby("seller_id")["in_other_state"].sum().reset_index() #в другие штаты

counts["share"] = other_state_counts["in_other_state"] / counts["order_id"] #доля

merged_data = counts[counts["order_id"] > 20] #продавцы с заказами больше 20 штук

merged_data = merged_data.sort_values(by="share", ascending=False).head(5) #от большего к меньшему и всего 5, знаю, что со срезами у пандаса проблемы, так что юзаем head

merged_data = merged_data[["seller_id", "share"]] #оставляем нужное


#гениально я считаю
merged_data_full=merged_data.merge(merged_data_full, how='left')
merged_data_full = (
    merged_data_full
    .merge(agg_location, left_on="seller_zip_code_prefix", right_on="geolocation_zip_code_prefix", how='left')
    .rename(columns={"geolocation_lat": "seller_lat", "geolocation_lng": "seller_lng"})
    .drop(columns=["geolocation_zip_code_prefix"]) #удаляем
    .merge(agg_location, left_on="customer_zip_code_prefix", right_on="geolocation_zip_code_prefix", how='left')
    .rename(columns={"geolocation_lat": "customer_lat", "geolocation_lng": "customer_lng"})
    .loc[:, ["seller_id", "customer_zip_code_prefix", "seller_lat", "seller_lng", "customer_lat", "customer_lng"]] # вот эту штуку подсказал сам colab
)
#merged_data_full = merged_data_full[["seller_id", "seller_lat", "seller_lng","customer_lat", "customer_lng"]] # вместо этой

seller_colors = {seller_id: color for seller_id, color in zip(merged_data['seller_id'], px.colors.qualitative.Plotly)} # украл у умных людей


fig = go.Figure()
for _, row in merged_data_full.iterrows():
    fig.add_trace(go.Scattermapbox(mode = "lines",
                              lat=[row["seller_lat"], row["customer_lat"]],
                              lon=[row["seller_lng"], row["customer_lng"]],
                              line={'color': seller_colors[row['seller_id']]}))
fig.update_layout(mapbox_style="open-street-map")
fig.show()
```
